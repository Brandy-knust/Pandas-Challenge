## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#pymoli_df.describe()  Number of Unique Items	Average Price	Number of Purchases	Total Revenue
#find the number of unique items
unique_items = pymoli_df["Item Name"].unique()
unique_items = len(unique_items)
unique_items
#find the average price
price = pymoli_df["Price"].mean()
price
#find the number of purchases
purchases = pymoli_df["Purchase ID"].count()
purchases
#find the total revenue
revenue = pymoli_df["Price"].sum()
revenue
#create a summary table
summary_df = pd.DataFrame({"Number of Unique Items":[unique_items],"Average Price":[price],"Number of Purchases":[purchases],"Total Revenue":[revenue]})
summary_df["Average Price"] = summary_df["Average Price"].astype(float).map("${:,.2f}".format)
summary_df["Total Revenue"] = summary_df["Total Revenue"].astype(float).map("${:,.2f}".format)
summary_df.head()
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#group by genders
purchases_gender = pymoli_df.groupby("Gender")
#count purchases
purchases_count = purchases_gender["Price"].count()
purchases_count 
#find the average purchase price and totals
purchase_price = purchases_gender["Price"].mean()
Total_value = purchases_gender["Price"].sum()
#find the percentage
purchases_gendercount = purchases_gender["SN"].nunique()
ATPPP = Total_value/purchases_gendercount
#create the summary table
purchases_gender_df = pd.DataFrame({"Purchase Count":purchases_count, "Average Purchase Price":purchase_price, "Total Purchase Value":Total_value, "Avg Total Purchase per Person":ATPPP})
#format the summary data frame
purchases_gender_df["Average Purchase Price"] = purchases_gender_df["Average Purchase Price"].astype(float).map("${:,.2f}".format)
purchases_gender_df["Total Purchase Value"] = purchases_gender_df["Total Purchase Value"].astype(float).map("${:,.2f}".format)
purchases_gender_df["Avg Total Purchase per Person"] = purchases_gender_df["Avg Total Purchase per Person"].astype(float).map("${:,.2f}".format)
purchases_gender_df

```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
#print(pymoli_df["Age"].max())
#print(pymoli_df["Age"].min())
#bin the data based on previously found min and max data
bins = [0, 9, 14, 19, 24, 29, 34, 39, 100]
age_labels = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
pymoli_df["Age_Groups"] = pd.cut(pymoli_df["Age"], bins, labels=age_labels)
```

```python
#create groups by age
age_group = pymoli_df.groupby("Age_Groups")
#find counts without the duplicates
age_group_count = age_group["SN"].nunique()
#calculate the total players and the percentages
age_group_total = age_group_count.sum()
age_group_percentage = age_group_count/age_group_total*100
#create the summary table
age_group_df = pd.DataFrame({"Total Count":age_group_count, "Percentage of Players":age_group_percentage})
#format the data
age_group_df["Percentage of Players"] = age_group_df["Percentage of Players"].astype(float).map("{:,.2f}%".format)
age_group_df
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#pull the previously binned data by age group
pur_age_group = pymoli_df.groupby("Age_Groups")
#count by item ID
pur_age_count = pur_age_group["Item ID"].count()
#find purchase price average
pur_age_price = pur_age_group["Price"].mean()
#find the total purchase value by age groups
pur_tot_age = pur_age_group["Price"].sum()
#find the average total by person (limiting to the individuals and not double counting)
pur_un_age = pur_age_group["SN"].nunique()
pur_ave_age = pur_tot_age/pur_un_age

#create the summary table
pur_age_df = pd.DataFrame({"Purchase Count":pur_age_count, "Average Purchase Price":pur_age_price, "Total Purchase Value":pur_tot_age, "Avg Total Purchase per Person":pur_ave_age})
#format for cleaner data
pur_age_df["Average Purchase Price"] = pur_age_df["Average Purchase Price"].astype(float).map("${:,.2f}".format)
pur_age_df["Total Purchase Value"] = pur_age_df["Total Purchase Value"].astype(float).map("${:,.2f}".format)
pur_age_df["Avg Total Purchase per Person"] = pur_age_df["Avg Total Purchase per Person"].astype(float).map("${:,.2f}".format)
pur_age_df
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#Purchase Count
#Average Purchase Price
#Total Purchase Value
#Top 5
#Group by the user id and find the count of each user
top_sp_group = pymoli_df.groupby("SN")
top_sp_count = top_sp_group["Price"].count()
#find the sum of each user
top_sp_total=top_sp_group["Price"].sum()
#find the Average purchase price of each spender
top_sp_APP = top_sp_total/top_sp_count
#create the summary table
top_spender_df = pd.DataFrame({"Purchase Count":top_sp_count, "Average Purchase Price":top_sp_APP, "Total Purchase Value":top_sp_total})

#format the data in the summary table to look better
top_spender_df["Average Purchase Price"] = top_spender_df["Average Purchase Price"].astype(float).map("${:,.2f}".format)
top_spender_df["Total Purchase Value"] = top_spender_df["Total Purchase Value"].astype(float).map("${:,.2f}".format)
#sort descending and print only the top 5
top_sp_sorted = top_spender_df.sort_values(["Total Purchase Value"], ascending=False)
top_sp_sorted.head()

```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#Find only the Item ID, Item Name, and Price
most_pop_df = pymoli_df[["Item ID", "Item Name", "Price"]]
#Group by both ID and Name
most_pop_group = most_pop_df.groupby(["Item ID", "Item Name"])
#count how many of each
most_pop_count = most_pop_group["Price"].count()
most_pop_count
#Find the total price value of each item
most_pop_pur_price = most_pop_group["Price"].sum()
most_pop_pur_price
#find the average item price of each item
ave_item_price = most_pop_group["Price"].mean()
ave_item_price
#create the summary table
most_poptotals_df = pd.DataFrame({"Purchase Count":most_pop_count, "Average Item Price":ave_item_price, "Total Purchase Value":most_pop_pur_price})
#format the data to look cleaner
most_poptotals_df["Average Item Price"] = most_poptotals_df["Average Item Price"].astype(float).map("${:,.2f}".format)
most_poptotals_df["Total Purchase Value"] = most_poptotals_df["Total Purchase Value"].astype(float).map("${:,.2f}".format)
#sort in descending order and display only the top five most popular items
most_poptotals_sorted_df=most_poptotals_df.sort_values(["Purchase Count"], ascending=False)
most_poptotals_sorted_df.head()
#most_poptotals_sorted_df["Average Item Price"]'${:, .2f}'.format
#pd.options.display.float_format = '{:, .2f}'.format
#most_poptotals_sorted_df
```
